# Limpieza y preparación de datos

En esta lectura revisaremos por qué la limpieza y la preparación de datos son pasos esenciales antes de analizar información o entrenar modelos. Un conjunto de datos puede verse "completo" a primera vista y aun así contener problemas que cambian conclusiones, distorsionan indicadores o deterioran el rendimiento de un modelo.

## ¿Por qué preparar datos?

En analítica de datos rara vez trabajamos con información perfectamente lista. Antes de responder preguntas o construir modelos debemos verificar si los datos representan adecuadamente la realidad que queremos estudiar. Preparar datos significa revisar, corregir, transformar y documentar el conjunto de datos para que sea confiable y útil.

Un proceso de preparación bien hecho ayuda a:

* Reducir errores de interpretación.
* Evitar que registros defectuosos dominen el análisis.
* Aumentar la trazabilidad del proceso analítico.
* Facilitar que otras personas entiendan y reutilicen el dataset.
* Mejorar la estabilidad de modelos predictivos y descriptivos.

## Valores faltantes

Los valores faltantes aparecen cuando una observación no tiene información para una o varias variables. En una base de clientes, por ejemplo, puede faltar la edad; en una encuesta, puede faltar la respuesta a una pregunta; en sensores, puede faltar una medición por una falla de captura.

No todos los faltantes significan lo mismo. A veces el dato no existe, a veces no se capturó, y a veces se ocultó deliberadamente. Por eso, antes de llenar o eliminar valores faltantes, conviene preguntarse:

* ¿La ausencia es esperada o es un error?
* ¿Cuántos registros están afectados?
* ¿La variable es crítica para el análisis?
* ¿El patrón de ausencia se concentra en algún grupo?

Algunas estrategias comunes para manejar valores faltantes son:

* Eliminar filas o columnas cuando la ausencia es pequeña y no afecta demasiado.
* Imputar con un valor simple, como media, mediana o moda.
* Crear una categoría explícita como `"Desconocido"` en variables categóricas.
* Estimar valores faltantes mediante reglas de negocio o modelos.

La mejor opción depende del contexto. Imputar sin pensar puede introducir sesgos.

## Duplicados

Un duplicado es un registro repetido parcial o totalmente. Puede originarse por errores de captura, integraciones de fuentes distintas o procesos que vuelven a cargar la misma información.

Los duplicados afectan conteos, promedios, tasas y frecuencias. Por ejemplo, si una venta aparece dos veces, el ingreso total se sobreestima. En otros casos el problema es más sutil: un mismo cliente puede estar repetido con pequeñas variaciones en el nombre o en la ciudad.

Por eso es útil distinguir entre:

* Duplicados exactos: todas las columnas coinciden.
* Duplicados por llave: se repite un identificador que debería ser único.
* Duplicados semánticos: representan la misma entidad pero con diferencias menores en el texto.

## Valores atípicos

Un valor atípico es una observación muy alejada del comportamiento general de la variable. No siempre es un error: podría ser un caso real e importante. Sin embargo, sí es una señal de que vale la pena revisar.

Para detectar valores atípicos suelen usarse reglas como:

* Revisar rangos válidos definidos por el negocio.
* Comparar con percentiles o con el rango intercuartílico (IQR).
* Apoyarse en visualizaciones como histogramas o diagramas de caja.

Si una base de edades contiene el valor `150`, probablemente hay un error. Si una base de ingresos contiene un valor muy alto, podría ser un cliente especial o una captura incorrecta. La decisión siempre necesita contexto.

## Calidad de datos

La calidad de datos es el grado en que un conjunto de datos sirve para el uso que le queremos dar. No es una propiedad absoluta: un dataset puede ser suficiente para una descripción general y al mismo tiempo insuficiente para un modelo de predicción delicado.

Algunas dimensiones frecuentes de calidad son:

* **Completitud:** qué tanto faltan datos.
* **Validez:** si los datos cumplen reglas de formato o rango.
* **Consistencia:** si no se contradicen entre sí.
* **Unicidad:** si no hay duplicados indebidos.
* **Actualidad:** si la información está vigente.
* **Precisión:** si representa adecuadamente la realidad medida.

## Flujo recomendado de preparación

Un flujo simple y útil para preparar un dataset puede ser:

**1.** Cargar y entender el origen de los datos.  
**2.** Revisar tipos de variables, tamaños y estructura.  
**3.** Identificar faltantes, duplicados e inconsistencias.  
**4.** Detectar posibles valores atípicos.  
**5.** Aplicar reglas de limpieza justificadas.  
**6.** Verificar cómo cambió el dataset después de limpiar.  
**7.** Documentar las decisiones tomadas.

## Idea clave

Preparar datos no es "arreglar por arreglar". Es tomar decisiones informadas para que el conjunto de datos sea coherente con la pregunta analítica. Un buen análisis no empieza con el modelo: empieza con datos entendibles, verificables y confiables.
