# Lectura y exploración de datasets

En esta lectura presentamos ideas básicas para cargar datasets, inspeccionar su estructura y resumir su contenido con estadísticas descriptivas. Estas tareas suelen ser el primer contacto real con una base de datos y orientan las decisiones analíticas posteriores.

## Lectura de datasets

Leer un dataset no significa únicamente abrir un archivo. También implica entender su formato, su estructura y las reglas mínimas necesarias para interpretarlo correctamente.

Entre los formatos más frecuentes en analítica se encuentran:

* `CSV`: simple y ampliamente usado.
* `Excel`: común en entornos organizacionales.
* `JSON`: frecuente en APIs y servicios web.
* `DTA`, `SAV`, `Parquet` y otros formatos especializados.

Cada formato trae retos distintos: codificación, separadores, nombres de hojas, tipos de datos, fechas o valores perdidos.

Algunas preguntas útiles al cargar un dataset son:

* ¿Cuántas filas y columnas tiene?
* ¿Qué representa cada fila?
* ¿Qué significa cada columna?
* ¿Qué variables deberían ser numéricas, de texto o de fecha?
* ¿Hay una llave que identifique cada registro?

## Exploración inicial de datos

Una vez leída la base de datos, conviene realizar una revisión rápida para identificar su forma general. Explorar no es todavía modelar ni limpiar en profundidad; es reconocer el terreno.

La exploración inicial suele incluir:

* Observar las primeras filas.
* Revisar nombres de columnas.
* Verificar tipos de datos.
* Medir cantidad de registros y variables.
* Identificar valores faltantes visibles.
* Reconocer variables categóricas y numéricas.

## Estadísticas descriptivas básicas

Las estadísticas descriptivas resumen el comportamiento de los datos sin construir todavía explicaciones causales o predicciones. Su objetivo es ofrecer una primera imagen cuantitativa del dataset.

En variables numéricas, algunas medidas frecuentes son:

* **Media:** promedio aritmético.
* **Mediana:** valor central.
* **Mínimo y máximo:** extremos observados.
* **Desviación estándar:** dispersión alrededor de la media.
* **Cuartiles:** puntos de referencia para entender la distribución.

En variables categóricas interesa más revisar:

* Frecuencias absolutas.
* Frecuencias relativas o porcentajes.
* Categorías más y menos comunes.

Por ejemplo, si una variable `canal_venta` tiene categorías `Tienda`, `Web` y `App`, una tabla de frecuencias permite entender de inmediato cuál canal domina.

## Explorar es hacer preguntas

Una buena exploración no consiste solo en ejecutar métodos de software. También supone formular preguntas sencillas:

* ¿Qué categorías aparecen?
* ¿Qué ciudad concentra más ventas?
* ¿Qué rango de edades es más frecuente?
* ¿La variable parece razonable o tiene valores sospechosos?
* ¿Hay diferencias marcadas entre grupos?

## Buenas prácticas

Al explorar datos conviene:

* Documentar el origen del archivo.
* Conservar una copia sin modificar.
* Anotar supuestos y hallazgos tempranos.
* No confundir una estadística descriptiva con una conclusión definitiva.
* Combinar números con inspección visual cuando sea posible.

## Idea clave

Leer y explorar un dataset es el punto de partida para cualquier análisis serio. Si esta etapa se hace con cuidado, es más fácil detectar problemas, entender oportunidades y decidir qué técnicas conviene aplicar después.
